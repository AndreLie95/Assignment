```python
%matplotlib inline
import random
import matplotlib.pyplot as plt
```

```python
# buatlah list untuk menampung head dan tail
head_tail = [0,0]

#loop sebanyak 50.000 trials
trials = 50000
trial = 0

while trial < trials:
    trial = trial + 1
    # memperoleh random angka 0 dan 1
    toss = random.randint(0,1)
    head_tail[toss] = head_tail[toss]+1

print(head_tail)

# show in chart
plt.pie(head_tail, autopct='%.2f%%', labels=('Head', 'Tail'))
```
